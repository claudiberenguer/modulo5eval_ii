#### Procesado de outlayers con HDBSCAN. Dataset resultante es `dset_cp_cat_minmax_filt`
- para quitar los outliers se decide hacer un clustering con HDBSCAN. Se escoge HDBSCAN porqué no asume un número de clusters y es un algoritmo que tiene en cuenta la existencia de outliers de forma natural (no alcanzables). HDBSCAN tiene la ventaja sobre DBSCAN que ajsta el epsilon automáticamente.
- el algoritmo necesita que el rango de las componentes no sea muy distinto, ya que sino unas dominarían, por lo que se hace un escalado. Para categorías el MinMaxScaler es más adecuado; de hecho ya se ha visto arriba que la estadística Z (que a fin de cuentas es el StandardScaler) no se obtienen buenos resultos. Aun así se va a probar con ambas. Para esablecer un valor inicial de epsilon se calcula la distancia media entre un grupo de instancias tomadas al azar y se usa la mitad de esa distancia (ver más abajo). Se demuestra que el clustering con normalización StandardScaler no da buen resultado, con MinMaxScaler, sí
- El número mínimo de elementos en el cluster se establece en 5 (hay algunas variables con valores normales pero con poca frecuencia) y el epsilon incial
- Los outliers se van a detectar por el label_ que resulte para la instancia:
      labels_ : ndarray of shape (n_samples,)
            Cluster labels for each point in the dataset given to :term:`fit`.
            Outliers are labeled as follows:
            - Noisy samples are given the label -1.
            - Samples with infinite elements (+/- np.inf) are given the label -2.
            - Samples with missing data are given the label -3, even if they also have infinite elements.
- Se descartan 140 de las 119.390 instancias (en una ejecución previa fueron 181). Se fija randomst
- El resultado escalado es el que se va a usar para el entrenamiento, ya que de los métodos a probar las redes neuronales lo requieren y en logistic regression también es conveniente
- el dataset resultante es `dset_cp_cat_minmax_filt`

```python

scaler_minmax = MinMaxScaler()
std_scaler = StandardScaler()

std_scaled = std_scaler.fit_transform(dset_cp_cat)
dset_cp_cat_std = pd.DataFrame(std_scaled, 
                                  columns=list(dset_cp_cat.columns))
minmax_scaled = scaler_minmax.fit_transform(dset_cp_cat)
dset_cp_cat_minmax = pd.DataFrame(minmax_scaled,
                                  columns=dset_cp_cat.columns)
```

```python
def average_dist(array, num_rows):
    np.random.shuffle(array)
    subset = array[:num_rows,:]
    avg_dist = 0
    for vec in subset:
        vec_mat = np.tile(vec, (num_rows,1))
        dif_mat = subset - vec_mat
        sq_mat = np.matmul(dif_mat, dif_mat.T)
        avg_d2 = np.average(sq_mat)
        avg_d = np.sqrt(avg_d2)
        avg_dist += avg_d/(num_rows -1)
    return avg_dist
```

```python
print(f'std_scaled average distance: {average_dist(std_scaled, 200):.2f}')
print(f'minmax_scaled average distance: {average_dist(minmax_scaled, 200):.2f}')
```

```python
def hdbscan(epsilon):
    return HDBSCAN(min_cluster_size=5,
                   cluster_selection_epsilon=epsilon,
                   metric='euclidean',
                   algorithm='auto',
                   n_jobs=-1,
                   cluster_selection_method='eom')
```

### 1.3 Preproceado de los datos (se genera el train_test_split, la transformación i el filtrado de outlayers en X_train)

```python
def loader(dataset_csv_filename:str) -> (pd.core.frame.DataFrame, pd.core.frame.DataFrame, pd.core.series.Series, pd.core.series.Series):
    
    # se setea random_state para que los resultados sean predecibles (que el resultado sea el mismo siempre)
    check_random_state(42)

    # carga del csv en un dataframe
    dset = pd.read_csv(DATASET)
    
    # eliminación de las columnas conflictivas
    dset.drop(['reservation_status', 'reservation_status_date'], axis=1, inplace=True)
    dset.drop(['company'], axis=1, inplace=True)
    dset.drop(['agent'], axis=1, inplace=True)
    dset_cp.drop(['arrival_date_year', 'arrival_date_day_of_month', 'arrival_date_month'], axis=1, inplace=True)

    # relleno de los valores faltantes
    dset['children'] = dset_cp['children'].fillna(dset_cp['children'].median())
    dset['country'] = dset_cp['country'].fillna('unknown')

    # conversión de los campos de texto en numéricos
    text_fields = [c for c in dset.columns if dset[c].dtype==object]
    for c in text_fields:
        dset[c] = dset[c].astype('category').cat.codes

    # separación en target y features
    y = dset['is_canceled']
    X = dset.drop(['is_canceled'], axis=1)

    # test - train split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  
    
    # escalado
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # eliminación de outlayers del conjunto de train
    hdbscan = HDBSCAN(min_cluster_size=5,
                      cluster_selection_epsilon=0.6,
                      metric='euclidean',algorithm='auto',
                      n_jobs=-1,
                      cluster_selection_method='eom')
    clusters = hdbscan.fit(X_train)
    X_train['cluster'] = clusters.labels_
    X_train = X_train[X_train.cluster!=-1].drop(['cluster'], axis=1)   

    return (X_train, X_test, y_train, y_test)
```

### 3.1 Justificación de la métrica escogida
Se escoge el `accuracy score`, dado que se trata de un dataset balanceado y en el que tiene tanto valor el detectar el caso positivo como el caso negativo. 

Si fuese dataset desbalanceado y donde fuese importante detectar bien los positivos, con un compromiso entre precision y recall, sería mejor usar `F1 score`

En caso de que lo importante fuese no dejarse ningun positivo a un a cuesta de tener algunos falsos positivos, se escogería `recall score`

En caso de que lo importante fuese evitar falsos positivos, aun a custa de tener algún falso negativo, se escogería `precision score`
